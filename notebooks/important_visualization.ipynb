```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plot Distribution of Insurance Premiums
plt.figure(figsize=(12, 6))
sns.histplot(data['TotalPremium'], kde=True, bins=30, color='skyblue')
plt.title('Distribution of Insurance Premiums')
plt.xlabel('Total Premium')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Premium vs. Claims Scatter Plot
plt.figure(figsize=(12, 6))
sns.scatterplot(x='TotalPremium', y='TotalClaims', data=data, color='orange', alpha=0.5)
sns.regplot(x='TotalPremium', y='TotalClaims', data=data, scatter=False, color='blue')
plt.title('Total Premium vs. Total Claims')
plt.xlabel('Total Premium')
plt.ylabel('Total Claims')
plt.grid(True)
plt.show()
```

```python
# Plot Distribution of Insurance Coverage Types by Country
# Group by country and coverage type, then count occurrences
coverage_distribution = data.groupby(['Country', 'CoverType']).size().unstack().fillna(0)
coverage_distribution.plot(kind='bar', stacked=True, figsize=(14, 8), colormap='tab20')
plt.title('Distribution of Insurance Coverage Types by Country')
plt.xlabel('Country')
plt.ylabel('Number of Policies')
plt.legend(title='Coverage Type')
plt.grid(True)
plt.show()
```
